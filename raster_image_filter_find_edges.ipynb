```python
import numpy as np
from PIL import Image, ImageFilter
from matplotlib import pyplot as plt
```

```python
filename="house.png"
img = Image.open(filename)
```

```python
edges = img.filter(ImageFilter.FIND_EDGES)
```

```python
plt.figure(1, figsize=(6,6), dpi=100)
plt.imshow(edges)
```

```python
plt.show()
```
