## Zobrazení objektů typu `java.awt.image.BufferedImage`

Nejprve je nutné provést import příslušného balíčku s využitím standardního makra `import`

```clojure
(import java.awt.image.BufferedImage)
```

Deklarace funkcí pro vytvoření nové bitmapy, funkce pro výpočet celé hodnoty představující barvu pixelu a funkce pro vybarvení zvoleného pixelu v bitmapě

```clojure
(defn create
  "Create new bitmap."
  [width height]
  (new BufferedImage width height BufferedImage/TYPE_INT_RGB))

(defn ->rgb
  [r g b]
  (bit-or b
          (bit-shift-left g 8)
          (bit-shift-left r 16)))

(defn putpixel
  "Change pixel color."
  ( [bitmap x y rgb]
    (.setRGB bitmap x y rgb))
  ( [bitmap x y r g b]
    (.setRGB bitmap x y (->rgb r g b))))
```

Hodnota vrácená funkcí `create` je typu `java.awt.image.BufferedImage` a je tedy Jupyter Notebookem přímo zobrazena

```clojure
(create 256 256)
```

Ukázka vybarvení všech pixelů v bitmapě

```clojure
(let [bitmap (create 256 256)]
    (doseq [y (range 256)]
        (doseq [x (range 256)]
            (putpixel bitmap x y x 255 y)))
    bitmap)
```
