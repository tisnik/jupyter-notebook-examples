```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
WIDTH = 256
HEIGHT = 256
raster = np.zeros(shape=(HEIGHT, WIDTH, 3), dtype=np.uint8)
```

```python
for y in range(HEIGHT):
    for x in range(WIDTH):
        raster[y][x][0] = x
        raster[y][x][1] = 255
        raster[y][x][2] = y
```

```python
plt.figure(1, figsize=(8,6), dpi=100)
plt.imshow(raster)
```

```python
plt.show()
```
