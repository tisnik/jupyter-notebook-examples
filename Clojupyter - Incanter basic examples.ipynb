# Otestování základní funkcionality knihoven z projektu Incanter

Nejprve je nutné provést import všech balíčků

```clojure
(use '(incanter core stats charts io))
```

Otestujeme, zda je možné zobrazit histogram pro pseudonáhodně generované hodnoty

```clojure
(-> (sample-normal 10000)
    histogram
    (.createBufferedImage 600 400))
```

Odlišný typ grafu

```clojure
(-> (scatter-plot (sample-normal 1000) 
                  (sample-normal 1000)
                  :x-label "x" :y-label "y")
    (.createBufferedImage 600 400))
```
