```python
import numpy as np
from PIL import Image, ImageFilter
from matplotlib import pyplot as plt
```

```python
filename="house.png"
img = Image.open(filename)
```

```python
blurred = img.filter(ImageFilter.BLUR)
```

```python
plt.figure(1, figsize=(8,6), dpi=100)
plt.imshow(blurred)
```

```python
plt.show()
```
